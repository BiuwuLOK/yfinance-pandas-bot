Setup:

Defines the stock ticker symbol (NVDA).

Defines fast and slow intervals for Simple Moving Averages (SMAs) (10 and 30).

Initializes a flag (currently_holding) to track if the asset is currently held and an empty list (tradelog) for logging trades.

Main Loop:

Continuously runs in an infinite while loop with a sleep interval of 60 seconds (time.sleep(60)).

Data Fetching:

Fetches the most recent 2 days of historical data for the AMZN ticker, with a 1-minute interval.

Removes unnecessary columns (Dividends, Stock Splits, Volume).

Calculate Technical Indicators:

Adds SMA_fast (10-period) and SMA_slow (30-period) moving averages of the closing prices to the DataFrame.

Trading Logic:

Buy Signal: If SMA_fast > SMA_slow and the asset is not currently held, it prints a buy message and logs the trade with datetime, ticker, side, and price. Sets currently_holding to True.

Sell Signal: If SMA_fast

```python
import time
import yfinance as yf
from datetime import datetime, timedelta
import pandas_ta as ta

ticker = "NVDA"
asset = yf.Ticker(ticker)
interval_fast = 10
interval_slow = 30
currently_holding = False
tradelog = []
```

```python
try:
  while True:
    # setting up real-timeline in past 3 months up to now
    start_date = (datetime.now() - timedelta(days=90)).strftime('%Y-%m-%d')
    end_date = datetime.now().strftime('%Y-%m-%d')

    # dataframe
    df = asset.history(start=start_date, end=end_date, interval='1d')
    del df['Dividends']
    del df['Stock Splits']
    del df['Volume']

    # Calculate the Simple Moving Averages (SMAs)
    df['SMA_fast'] = ta.sma(df['Close'], interval_fast) # adv fast
    df['SMA_slow'] = ta.sma(df['Close'], interval_slow) # adv slow
    price = df.iloc[-1]['Close']

    # main logic
    if df.iloc[-1]['SMA_fast'] > df.iloc[-1]['SMA_slow'] and not currently_holding:
        print(f"Buy {ticker}@{price}")
        tradelog.append({'date':datetime.now(),'ticker': ticker,'side': 'buy','price': price})
        currently_holding = True

    elif df.iloc[-1]['SMA_fast'] < df.iloc[-1]['SMA_slow'] and currently_holding:
        print(f"Sell {ticker}@{price}")
        tradelog.append({'date':datetime.now(),'ticker': ticker,'side': 'sell','price': price})
        currently_holding = False

    time.sleep(60) # work each 60s
except KeyboardInterrupt:
  print("Exucution stopped.")
```

```python
# Record the start time
start_time = time.time()
# Duration for which to run the loop (10 minutes = 600 seconds)
duration = 600

try:
  while True:
      # Check if 10 minutes have passed
      elapsed_time = time.time() - start_time
      if elapsed_time > duration:
          print("Program completed")
          break

      # Set the date range to the past 3 months up to now
      start_date = (datetime.now() - timedelta(days=90)).strftime('%Y-%m-%d')
      end_date = datetime.now().strftime('%Y-%m-%d')

      # Fetch the historical data with a daily interval
      df = asset.history(start=start_date, end=end_date, interval='1d')
      del df['Dividends']
      del df['Stock Splits']
      del df['Volume']

      # Calculate the Simple Moving Averages (SMAs)
      df['SMA_fast'] = ta.sma(df['Close'], interval_fast)
      df['SMA_slow'] = ta.sma(df['Close'], interval_slow)

      # Trading logic
      price = df.iloc[-1]['Close']
      if df.iloc[-1]['SMA_fast'] > df.iloc[-1]['SMA_slow'] and not currently_holding:
          print(f"Buy {ticker}@{price}")
          tradelog.append({'date':datetime.now(),'ticker': ticker,'side': 'buy','price': price})
          currently_holding = True

      elif df.iloc[-1]['SMA_fast'] < df.iloc[-1]['SMA_slow'] and currently_holding:
          print(f"Sell {ticker}@{price}")
          tradelog.append({'date':datetime.now(),'ticker': ticker,'side': 'sell','price': price})
          currently_holding = False

      # Wait for 1 minute before the next iteration
      time.sleep(60)

  # Print out the trade log at the end of the program
  print("\nTrade Log:")
  for trade in tradelog:
      print(trade)

except KeyboardInterrupt:
  print("Exucution stopped.")
```

Multi-asset trading

```python
import time
import yfinance as yf
from datetime import datetime, timedelta
import pandas_ta as ta

# List of assets to trade
asset_list = [
    {'ticker': 'AMZN', 'yfticker': yf.Ticker('AMZN'), 'holding': False},
    {'ticker': 'NVDA', 'yfticker': yf.Ticker('NVDA'), 'holding': False},
    {'ticker': 'GOOG', 'yfticker': yf.Ticker('GOOG'), 'holding': False},
    {'ticker': 'NFLX', 'yfticker': yf.Ticker('NFLX'), 'holding': False},
    {'ticker': 'MSFT', 'yfticker': yf.Ticker('MSFT'), 'holding': False},
    {'ticker': 'TSLA', 'yfticker': yf.Ticker('TSLA'), 'holding': False},
    {'ticker': 'BTC-USD', 'yfticker': yf.Ticker('BTC-USD'), 'holding': False},
    {'ticker': 'ETH-USD', 'yfticker': yf.Ticker('ETH-USD'), 'holding': False}
]

interval_fast = 10
interval_slow = 30
tradelog = []
```

```python
# Record the start time
start_time = time.time()
# Duration for which to run the loop (10 minutes = 600 seconds)
duration = 600

while True:
    try:
        # Check if 10 minutes have passed
        elapsed_time = time.time() - start_time
        if elapsed_time > duration:
            print("Program completed")
            break

        for asset in asset_list:
            ticker = asset['ticker']
            yfticker = asset['yfticker']
            currently_holding = asset['holding']

            # Set the date range to the past 3 months up to now
            start_date = (datetime.now() - timedelta(days=90)).strftime('%Y-%m-%d')
            end_date = datetime.now().strftime('%Y-%m-%d')

            # Fetch the historical data with a daily interval
            df = yfticker.history(start=start_date, end=end_date, interval='1d')
            del df['Dividends']
            del df['Stock Splits']
            del df['Volume']

            # Calculate the Simple Moving Averages (SMAs)
            df['SMA_fast'] = ta.sma(df['Close'], interval_fast)
            df['SMA_slow'] = ta.sma(df['Close'], interval_slow)

            # Check if both SMAs are not None
            if df['SMA_fast'].iloc[-1] is not None and df['SMA_slow'].iloc[-1] is not None:
                price = df.iloc[-1]['Close']
                if df.iloc[-1]['SMA_fast'] > df.iloc[-1]['SMA_slow'] and not currently_holding:
                    print(f"Buy {ticker}@{price}")
                    tradelog.append({'date': datetime.now(), 'ticker': ticker, 'side': 'buy', 'price': price})
                    asset['holding'] = True

                elif df.iloc[-1]['SMA_fast'] < df.iloc[-1]['SMA_slow'] and currently_holding:
                    print(f"Sell {ticker}@{price}")
                    tradelog.append({'date': datetime.now(), 'ticker': ticker, 'side': 'sell', 'price': price})
                    asset['holding'] = False

        # Wait for 1 minute before the next iteration
        time.sleep(60)

    except Exception as e:
        print(f"An error occurred: {e}")
        break

# Print out the trade log at the end of the program
print("\nTrade Log:")
for trade in tradelog:
    print(trade)
```

```python
import time
import yfinance as yf
from datetime import datetime, timedelta
import pandas_ta as ta
import math

# List of assets to trade
asset_list = [
    {'ticker': 'AMZN', 'yfticker': yf.Ticker('AMZN'), 'holding': False},
    {'ticker': 'NVDA', 'yfticker': yf.Ticker('NVDA'), 'holding': False},
    {'ticker': 'GOOG', 'yfticker': yf.Ticker('GOOG'), 'holding': False},
    {'ticker': 'NFLX', 'yfticker': yf.Ticker('NFLX'), 'holding': False},
    {'ticker': 'MSFT', 'yfticker': yf.Ticker('MSFT'), 'holding': False},
    {'ticker': 'TSLA', 'yfticker': yf.Ticker('TSLA'), 'holding': False},
    {'ticker': 'BTC-USD', 'yfticker': yf.Ticker('BTC-USD'), 'holding': False},
    {'ticker': 'ETH-USD', 'yfticker': yf.Ticker('ETH-USD'), 'holding': False}
]

def get_pause():
    now = datetime.now()
    next_min = now.replace(second=0, microsecond=0) + timedelta(minutes=1)
    pause = math.ceil((next_min - now).seconds)
    print(f"Sleep for {pause}")
    return pause

interval_fast = 10
interval_slow = 30
tradelog = []

# Record the start time
start_time = time.time()
# Duration for which to run the loop (10 minutes = 600 seconds)
duration = 600

while True:
    try:
        # Check if 10 minutes have passed
        elapsed_time = time.time() - start_time
        if elapsed_time > duration:
            print("Program completed")
            break

        for asset in asset_list:
            ticker = asset['ticker']
            yfticker = asset['yfticker']
            currently_holding = asset['holding']

            # Set the date range to the past 3 months up to now
            start_date = (datetime.now() - timedelta(days=90)).strftime('%Y-%m-%d')
            end_date = datetime.now().strftime('%Y-%m-%d')

            # Fetch the historical data with a daily interval
            df = yfticker.history(start=start_date, end=end_date, interval='1d')
            del df['Dividends']
            del df['Stock Splits']
            del df['Volume']

            # Calculate the Simple Moving Averages (SMAs)
            df['SMA_fast'] = ta.sma(df['Close'], interval_fast)
            df['SMA_slow'] = ta.sma(df['Close'], interval_slow)

            # Check if both SMAs are not None
            if df['SMA_fast'].iloc[-1] is not None and df['SMA_slow'].iloc[-1] is not None:
                price = df.iloc[-1]['Close']
                if df.iloc[-1]['SMA_fast'] > df.iloc[-1]['SMA_slow'] and not currently_holding:
                    print(f"Buy {ticker}@{price}")
                    tradelog.append({'date': datetime.now(), 'ticker': ticker, 'side': 'buy', 'price': price})
                    asset['holding'] = True

                elif df.iloc[-1]['SMA_fast'] < df.iloc[-1]['SMA_slow'] and currently_holding:
                    print(f"Sell {ticker}@{price}")
                    tradelog.append({'date': datetime.now(), 'ticker': ticker, 'side': 'sell', 'price': price})
                    asset['holding'] = False

        # Calculate the remaining seconds until the next minute starts and sleep for that duration
        time.sleep(get_pause())

    except Exception as e:
        print(f"An error occurred: {e}")
        break

# Print out the trade log at the end of the program
print("\nTrade Log:")
for trade in tradelog:
    print(trade)
```
